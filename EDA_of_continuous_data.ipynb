# EDA of continuous data

Summary: 

No evidence: chol, trestbps, ca.

Evidence: thalach, oldpeak, age.

```r
source("helpers.r")
library(tidyverse)
df <- get_training_df()
```

## `chol`: serum cholestoral in mg/dl

```r
summary(df$chol)
mean(df$chol, trim = 0.1)
```

```r
ggplot(df, aes(x = 0, y = df$chol))+
    geom_boxplot() + 
    geom_jitter(aes(x = 0, y = chol, color = target), width = 0.05, alpha = 0.5) +
    theme_bw()
```

```r
ggplot(df, aes(chol)) + 
    geom_density(kernel = "gaussian") + 
    theme_bw()
```

```r
ggplot(df, aes(log(chol))) + 
    geom_density(kernel = "gaussian") + 
    theme_bw()
```

```r
ggplot(df, aes(chol, color = target,  fill = target)) + 
    geom_density(kernel = "gaussian", alpha = 0.05) +
    theme_bw()
```

```r
ggplot(df)+
    geom_boxplot(aes(target, chol)) + 
    geom_jitter(aes(target, chol), width = 0.1, alpha = 0.5) +
    theme_bw() 
```

No evidence for `chol`.
