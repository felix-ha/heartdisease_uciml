# Simpson's Paradox

I was just curious if i can find Simpson's paradox in the data set. I did this for all dichotomous variables. Unfortunately (or fortunately) it does not occur. 

```r
source("helpers.r")

df <- get_training_df_clean()

predictors_base <- c("sex", "exang", "fbs")
predictors_granular <- c("sex", "exang", "fbs")

for(predictor_base in predictors_base) {
  for(predictor_granular in predictors_granular){
    
    if(predictor_granular == predictor_base) 
      next

  table_abs<- table(df[[predictor_base]], df$target)
  
  df_male <- df %>% filter((!!as.symbol(predictor_granular)) == levels(df[[predictor_granular]])[1])
  df_female <- df %>% filter((!!as.symbol(predictor_granular)) == levels(df[[predictor_granular]])[2])
  
  table_abs_male <- table(df_male[[predictor_base]], df_male$target)
  table_abs_female <- table(df_female[[predictor_base]], df_female$target)
  
  table_all <- prop.table(table_abs, 1)
  table_level_a <- prop.table(table_abs_male, 1)
  table_level_b <- prop.table(table_abs_female, 1)
  
  disease_all <- FALSE
  if(table_all[1,2] < table_all[2,2]) disease_all <- TRUE
  
  disease_level_a <- FALSE
  if(table_level_a[1,2] < table_level_a[2,2]) disease_level_a <- TRUE
  
  disease_level_b <- FALSE
  if(table_level_b[1,2] < table_level_b[2,2]) disease_level_b <- TRUE
  
  
  # print(disease_all)
  # print(disease_level_a)
  # print(disease_level_b)
  
  simpson <- FALSE
  if(disease_all == 0 && disease_level_a + disease_level_b == 2 || 
     disease_all == 1 && disease_level_a + disease_level_b == 0) simpson <- TRUE 
  
  print(simpson)

  }
}
```
